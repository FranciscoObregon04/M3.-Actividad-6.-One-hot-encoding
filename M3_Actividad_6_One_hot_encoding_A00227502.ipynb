**One hot encoding con pandas**

```python
#one hot encoding con pandas
titanicdf_nuevo = pd.get_dummies(titanicdf.Sex,prefix='Sex')
titanicdf_nuevo = titanicdf_nuevo.astype(int)
titanicdf_nuevo.head()
```

```python
titanicdf = titanicdf.drop(["Sex"],axis = 1)
titanicdf.shape
```

```python
titanicdf_onehotencoding = pd.concat([titanicdf,titanicdf_nuevo],axis=1)
titanicdf_onehotencoding.shape
#agregamos las columnas al dataframe titanic df_onehotencoding que se crearon en titanicdf_nuevo
```

```python
titanicdf_onehotencoding.head() #se puede ver que se elimino correctamente la columna de sexo y se agregaron dos columnas de femenino y masculino con valores numericos
```

**One hot encoding con scikit**




```python
titanicdf = pd.read_csv('/content/titanic.csv')
titanicdf.head()
```

```python
codificador = OneHotEncoder()

codificacion = codificador.fit_transform(titanicdf[["Sex"]])

#print(codificacion)
#hacemos el arreglo de la columna sex que es la categorica
print(codificacion.toarray())
```

```python
titanicdf.head()
```

```python
#creamos el datafarme en base al arreglo de la columna
nuevas_cols = pd.DataFrame(codificacion.toarray(),columns=codificador.categories_)
print(nuevas_cols)
```

```python
#concatenaremos al dataframe titanic df el dataframe nuevas_cols que contienen las nuevas columnas que se agregaron para masculino y femenino
titanicdf = pd.concat([titanicdf,nuevas_cols],axis="columns")
titanicdf
```

```python
#funcion para eliminar una columna del dataframe
titanicdf.drop("Sex",axis="columns",inplace=True)
titanicdf
```
